# Task: highest alcohol %? (quality wise)


```python
series= table_wine['quality'].value_counts()
```

```python
# bar chart of the distribution
import matplotlib.pyplot as plt
```

```python
y_axis=list(series)
x_axis=list(series.index)


plt.style.use('fivethirtyeight')

plt.bar(x_axis,y_axis)
plt.title('Alcohol quality distribution chart')
plt.xlabel('Quality')
plt.ylabel('Frequency')

```

```python
table_wine=table_wine.sort_values(by='alcohol',ascending=False)
```

```python
table_wine
```

```python
quality_grp=table_wine.groupby('quality')
```

```python
table_wine['rank']=quality_grp['alcohol'].rank('dense',ascending=False)
```

```python
table_wine[table_wine['rank']==1]
```

```python
data=table_wine[table_wine['rank']==1].drop_duplicates('quality')
data
```

```python
data[['alcohol','quality']]
```

```python
x_axis=data['quality']
y_axis=data['alcohol']

plt.bar(x_axis,y_axis,color='red',edgecolor='black')
plt.title('highest alcohol% ')
plt.xlabel('quality')
plt.ylabel('alcohol%')
plt.grid(True)
```
